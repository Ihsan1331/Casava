# 4. Sorting Dataset

```python
!mkdir sorted data_image
```

```python
label['label'].unique()
```

```python
!mkdir sorted/Cassava_Mosaic_Disease
!mkdir sorted/Cassava_Bacterial_Blight
!mkdir sorted/Cassava_Brown_Streak_Disease
!mkdir sorted/Cassava_Green_Mottle
!mkdir sorted/Cassava_Healthy
```

```python
CBB = []
CMD = []
CBSD = []
CGM = []
Healthy = []
for i in range (0, label.shape[0]):
  if 'CBB' in label.label[i]:
    CBB.append(label.image_id[i])
  elif 'CMD' in label.label[i]:
    CMD.append(label.image_id[i])
  elif 'CBSD' in label.label[i]:
    CBSD.append(label.image_id[i])
  elif 'CGM' in label.label[i]:
    CGM.append(label.image_id[i])
  else:
    Healthy.append(label.image_id[i])
```

```python
def move(source, destination):
  source_file = '/content/Data/train_images/'+ str(i)
  destination_folder = destination
  try:
    shutil.move(source_file, destination_folder)
  except FileNotFoundError:
    pass
```

```python
for i in CBB:
  move(i,'/content/sorted/Cassava_Bacterial_Blight')
```

```python
for i in CBSD:
  move(i,'/content/sorted/Cassava_Brown_Streak_Disease')
```

```python
for i in CGM:
  move(i,'/content/sorted/Cassava_Green_Mottle')
```

```python
for i in Healthy:
  move(i,'/content/sorted/Cassava_Healthy')
```

```python
for i in CMD:
  move(i,'/content/sorted/Cassava_Mosaic_Disease')
```
